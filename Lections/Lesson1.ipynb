# Задачи машинного обучения:
- Регрессия - прогноз на основе выборки объектов с различными признаками. На выходе -  вещественное число, то есть эта задача используется для оценки, которая имеет числовой эквивалент, например оценка стоимости акций, ожидаемый доход и т.д. В этой задаче множество Y (конечных состояний) - бесконечно
- Классификация - получение категориального ответа на основе набора признаков. Это значит, что проанализировав какой-то спектр параметров, мы выдаём ответ, к какой категории относится рассматриваемый объект. Количество ответов конечно (зачастую это да / нет - принадлежность к чему-либо), например является ли изображение лицом человека, можно ли одобрить кредит и т.д.
- Ранжирование
- Кластеризация
- Понижение размерности

# Методы машинного обучения с учителем

```python
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

Рассмотрим модель регрессии. Представлены множества исходных и конечных значений для линейного уравнения (k * x + b)

```python
x = np.array([4, 8, 9, 26])
y = np.array([39.2, 46.4, 48.2, 78.8])
```

```python
# Выводим график, дабы убедиться, что мы имеем линейную зависимость
plt.plot(x, y)
```

Создание модели линейной регрессии (LinearRegression). После этого производим обучение модели, для большинства моделей в `sklearn` за обучение отвечает метод `fit()`. На вход метода обучения подаётся обучающая выборка (множество объектов X, Y) - исходные и конечные значения

Зачастую, обучение модели может длиться долго, поэтому обученные модели сохраняют в битово. Скорость обучения зависит от модели, обучающей выборки и тд

P.S. Множество x было дополнено до двумерного массива (в обучающей выборке подразумевается набор параметров, а не только значение x), т.к. у нас есть только значения x, то для повышения размерности массива был использован `newaxis`

```python
one_x = 256

reg = linear_model.LinearRegression()
reg.fit(x[:, np.newaxis], y)
```

Коэффицент k линейной регрессии

```python
reg.coef_
```

Коэффицент b линейной регрессии (пересечение графика с осью Y)

```python
reg.intercept_
```

Для использования обученной модели на входных параметрах используется метод `predict(<value>)`, в последствии с помощью этого метода можно сравнить результаты обученной модели с проверочными данными

P.S. В качестве параметра, `predict` принимает массив входных значений

```python
reg.predict([[one_x]])
```
