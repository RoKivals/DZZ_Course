# Теория
Машинное обучение - исследование компьютерных алгоритмов, которые автоматически улучшаются благодаря опыту и использованию данных

# Машинное обучение
- С учителем (учитель поясняет, какие данные были верны на учебных данных)
- Без учителя (нет указаний на правильные ответы)
- Обучение с подкреплением (учителя нет, агент сам пытается определить правильность модели)

# Задачи машинного обучения:
- Регрессия
- Классификация
- Ранжирование
- Кластеризация
- Понижение размерности

Задача регрессии - прогноз на основе выборки объектов с различными признаками. На выходе -  вещественное число, то есть эта задача используется для оценки, которая имеет числовой эквивалент, например оценка стоимости акций, ожидаемый доход и т.д. В этой задаче множество Y (конечных состояний) - бесконечно

Задача классификации - получение категориального ответа на основе набора признаков. Это значит, что проанализировав какой-то спектр параметров, мы выдаём ответ, к какой категории относится рассматриваемый объект. Количество ответов конечно (зачастую это да / нет - принадлежность к чему-либо), например является ли изображение лицом человека, можно ли одобрить кредит и т.д.

# Методы машинного обучения с учителем

```python
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([4, 8, 9, 26])
y = np.array([39.2, 46.4, 48.2, 78.8])
```

```python
plt.plot(x, y)
```

```python
one_x = 256

reg = linear_model.LinearRegression()
reg.fit(x[:, np.newaxis], y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[one_x]])
```
